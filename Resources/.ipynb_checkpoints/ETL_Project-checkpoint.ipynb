```python
import pandas as pd
from sqlalchemy import create_engine

import sqlite3
```

```python
conn = sqlite3.connect('austinproject2.sqlite')
```

```python
c = conn.cursor()
conn.execute('drop table if exists aac_outcomes')

conn.execute('drop table if exists city_satisfaction')
```

```python
c.execute('''CREATE TABLE aac_outcomes
    (id INT PRIMARY KEY,
    year integer, 
    animal_type VARCHAR(30), 
    outcome_type VARCHAR(30),
    age_upon_outcome TEXT,
    breed TEXT)''')
          
```

```python
c.execute('''CREATE TABLE city_satisfaction
    (id INT PRIMARY KEY,
    year INT,
    quality_of_life VARCHAR(20),
    parks_and_rec VARCHAR(20),
    animal_services VARCHAR(20),
    safety_day VARCHAR(20),
    safety_night VARCHAR(20),
    safety_parks VARCHAR(20),
    num_parks VARCHAR(20),
    num_trails VARCHAR(20),
    park_appearance VARCHAR(20),
    age_range TEXT,
    income_range TEXT,
    zip_code integer)''')
```

```python
conn.commit()
```

## Extract the CSVs into DataFrames

```python
aac_outcomes_csv = "Austin_Animal_Center_Outcomes.csv"
aac_outcomes_df = pd.read_csv(aac_outcomes_csv)
aac_outcomes_df.head()
```

```python
city_satisfaction_csv = "Community_Survey_2015__2016___2017.csv"
city_satisfaction_df = pd.read_csv(city_satisfaction_csv)
city_satisfaction_df.head()
```

## Transform the DataFrames

```python
new_aac_outcomes_df = aac_outcomes_df[['ID','DateTime', 'Outcome Type', 'Animal Type', 'Breed', 'Age upon Outcome']].copy()
new_aac_outcomes_df = new_aac_outcomes_df.rename(index=str, columns={"ID": "id", "DateTime": "year", "Outcome Type": "outcome_type", "Animal Type": "animal_type", "Breed":"breed", "Age upon Outcome": "age_upon_outcome"})
new_aac_outcomes_df.set_index("id", inplace=True)
new_aac_outcomes_df['year']
# new_animals_df.drop_duplicates("id", inplace=True)
```

```python
new_aac_outcomes_df['year'] = pd.DatetimeIndex(new_aac_outcomes_df['year']).year
```

```python
new_aac_outcomes_df['year'].value_counts()
#we need to drop 2013, 2014, 2018, 2019
```

```python
new_aac_outcomes_df = new_aac_outcomes_df[new_aac_outcomes_df.year != 2013]
new_aac_outcomes_df = new_aac_outcomes_df[new_aac_outcomes_df.year != 2014]
new_aac_outcomes_df = new_aac_outcomes_df[new_aac_outcomes_df.year != 2018]
new_aac_outcomes_df = new_aac_outcomes_df[new_aac_outcomes_df.year != 2019]
```

```python
new_aac_outcomes_df['outcome_type'] = new_aac_outcomes_df['outcome_type'].replace('Disposal', 'Died')
```

```python
new_aac_outcomes_df['outcome_type'] = new_aac_outcomes_df['outcome_type'].replace('Rto-Adopt', 'Return to Owner')
```

```python
new_aac_outcomes_df.head()
```

```python
# new_aac_outcomes_df.groupby(['year','animal_type','outcome_type']).count()
```

```python
new_aac_outcomes_df.head()
```

```python
new_city_satisfaction_df = city_satisfaction_df[["ID", "Year", "Overall quality of life in the city", "Overall quality of city parks and recreation", "Animal services (shelter, adoptions, animal control, etc.)","I feel safe in my neighborhood during the day", "I feel safe in my neighborhood at night","I feel safe in city parks","Number of City of Austin parks","Number of City of Austin walking/biking trails","Appearance of City of Austin park grounds located in Austin","Which of the following best describes your AGE?","Which of the following best describes your annual household income?","Zip Code"]].copy()
new_city_satisfaction_df = new_city_satisfaction_df.rename(index=str, columns={"ID": "id", "Year": "year", "Overall quality of life in the city": "quality_of_life", "Overall quality of city parks and recreation":"parks_and_rec", "Animal services (shelter, adoptions, animal control, etc.)":"animal_services","I feel safe in my neighborhood during the day":"safety_day","I feel safe in my neighborhood at night":"safety_night","I feel safe in city parks":"safety_parks","Number of City of Austin parks":"num_parks","Number of City of Austin walking/biking trails":"num_trails","Appearance of City of Austin park grounds located in Austin":"park_appearance","Which of the following best describes your AGE?":"age_range","Which of the following best describes your annual household income?":"income_range","Zip Code":"zip_code"})
# new_city_satisfaction_df.drop_duplicates("id", inplace=True)
new_city_satisfaction_df.set_index("id", inplace=True)
new_city_satisfaction_df.head()
```
